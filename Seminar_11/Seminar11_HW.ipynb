```python
import numpy as np
import matplotlib.pyplot as plt
```

## 4. Построить график

```python
x = np.arange(-10, 10.01, 0.01)
func = -12*x**4 * (np.sin(np.cos(x))) - 18*x**3 + 5*x**2 + 10*x - 30

plt.plot(x, func)
plt.show()
```

## 6, 7. Определить промежутки, на котором f > 0 И f<0



```python
x = np.arange(-10, 10.01, 0.01)
func = -12*x**4 * (np.sin(np.cos(x))) - 18*x**3 + 5*x**2 + 10*x - 30

pos_list = []
neg_list = []

for i in x:
    res = -12*i**4 * (np.sin(np.cos(i))) - 18*i**3 + 5*i**2 + 10*i - 30
    if res > 0:
        pos_list.append(res)
    else:
        neg_list.append(res)

print(f'\n{pos_list[0]} --> {pos_list[-1]}\n')
print(pos_list)
print(f'\n{neg_list[0]} --> {neg_list[-1]}\n')
print(neg_list)
```
